# Questions to consider:

1. [Women and children first](https://en.wikipedia.org/wiki/Women_and_children_firstis) is a historical code of conduct whereby the lives of women and children were to be saved first in a life-threatening situation. Did this hold true on the Titanic?

2. Passenger class played an important role on the Titanic. Those in the upper class are more likely to survive given their influence and wealth. Did more passengers in the upper class survive? Did those in upper class pay more for their ticket?

3. Passengers embarked from Cherbourg, Queenstown and Southampton. How did this effect whether they survived or not?

4. Would having a large family on the Titanic be good for survival? 



```python
# I'll first import the necessary libraries, load the csv data, and get some quick info about the dataset. 
# I see Age and Embarked are missing some data points and less than half the cabin column exists.  

import pandas as pd
import numpy as np
import seaborn as sns

%pylab inline

df = pd.read_csv('titanic_data.csv')
df.info()
```

```python
# Looking at the above questions it's clear I won't need some columns for the analysis.
df = df.drop(['Name','Ticket','Cabin', 'PassengerId'],axis=1)        
```

```python
# Only 714 people have an age in the dataset. I'll make sure to add in the missing datapoints. 
df.Age = df.Age.interpolate()
```

```python
# Question 1: This question requires I look at Age and Gender. I'll take a closer look at Age first.
# Exploring the age data we find there's an average age of 30 years old with standard deviation of 14 years and the marjority of users were under 40.
df.Age.describe()
```

```python
# Let's take a closer look at who survived given their age.
age_survival = sns.FacetGrid(df, hue="Survived",aspect=3)
age_survival.map(sns.kdeplot,'Age',shade= True)
age_survival.set(xlim=(0, df['Age'].max()))
age_survival.fig.suptitle('Passenger Survival with respect to their Age') 
age_survival.add_legend()
```

```python
# We see that children(age < ~13) on aboard seem to have a higher chance of survival. Let's now factor in gender. 
# I'll start by creating a new column that seperates passenger into three groups - females, children and males. 
# As epxplained in the question. We are looking to see who has the higher rate of survival amoung these groups. 

def get_person(passenger):
    age,sex = passenger
    return 'child' if age < 13 else sex

# Create Person column to keep track of all 3 groups.
df['Person'] = df[['Age','Sex']].apply(get_person,axis=1)

survival_count = df[["Person", "Survived"]].groupby(['Person'],as_index=False).count() 
print survival_count

avg_survival = df[["Person", "Survived"]].groupby(['Person'],as_index=False).mean() 
print avg_survival

# From the data we can we can say that the majority of deaths were males over the age of 13. 
# 54% of the children and 76% of women over the age of 13 survived. 
```

```python
# We can better understand this data by visualizing it. 
# It's interesting that, on average, more women survived then children.  
fig, (axis1,axis2) = plt.subplots(1,2,figsize=(10,5))

# In the first graph I'll show how many women, men and children died on the Titanic. 
sns.countplot(x='Person', data=df, ax=axis1, palette="Blues_d")

# In the second graph I'll show what percentage of each group died the on the Titanic.
sns.barplot(x='Person', y='Survived', data=avg_survival, ax=axis2, order=['male','female','child'], palette="Blues_d")
```

```python
# Question #2: Did more passengers in the upper class survive? Did those in upper class pay more for their ticket?
# I'll start by defining a way to analyze survival with repsect to another passenger property.
def survival_analysis_of(prop):
    temp11 = df.groupby(prop).Survived.count()
    temp12 = df.groupby(prop).Survived.sum()/df.groupby(prop).Survived.count()
    print 'Total number of passengers'
    print temp11
    print '\nAverage number of passengers who survived'
    print temp12
    fig11 = plt.figure(figsize=(8,4))
    ax11 = fig11.add_subplot(121)
    ax11.set_xlabel(prop)
    ax11.set_ylabel('Count of Passengers')
    ax11.set_title("Passengers by {}".format(prop))
    temp11.plot(kind='bar')

    ax12 = fig11.add_subplot(122)
    temp12.plot(kind = 'bar')
    ax12.set_xlabel(prop)
    ax12.set_ylabel('Survival, on average, with respect to {}'.format(prop))
    ax12.set_title("Survival vs {}".format(prop))

# We see that, on average, the majority of people on the bottom class died on the Titanic which is what we expected. 
survival_analysis_of('Pclass')  
```

```python
# The second part of question 2 deals with the correlation between fare price and class. 
df = pd.read_csv('titanic_data.csv')
print 'Distribution of passenger fares:'
print df.Fare.describe()

def standardize(value):
    return (value - value.mean()) / value.std(ddof=0)

def correlation(x,y):
    return (standardize(x) * standardize(y)).mean()

# We see a inverse relationship between fare price and what passenger class. Those in a higher class tend to pay more. 
print "\nThe correlation between class and a passenger's fare price: {}".format(correlation(df.Fare,df.Pclass))

# I noticed some data was strange and seemed like an outlier. The passengers paid 0 in upper class whih doesnt seem right. 
# Looking at subset of the data for passengers who paid between $1 and less then $200 I found a strong correlation between fare and class. 
# This support our ealier intuition that passengers who paid less fare were in a lower class.  
df = pd.read_csv('titanic_data.csv')
print '\nDistribution of passenger fares excluding outliers:'
df = df[df.Fare >= 1][df.Fare < 200]
print df.Fare.describe()
print "\nThe correlation between class and a passenger's fare price excluding some outlying data: {}".format(correlation(df.Fare,df.Pclass))

# visualize fare distribution across each class
df.boxplot(column='Fare', by = 'Pclass')
plt.title("Distribution of fares for each class")
plt.suptitle("")
```

```python
# Question 3: What type of passengers came from where and how did this effect their survival?
# Using the same analysis we did for class we can take a closer look at survival with respect to where a passenger embarked from.
survival_analysis_of('Embarked') 
```

```python
# Question 4: Would having a large family be good for survival? 
# Instead of having two columns Parch & SibSp, we can have only one column that indicates wether a passanger was alone or not.
# We can them run an analysis of survival with respect to whether a passenger was alone or not. 
df['Family'] =  df['Parch'] + df['SibSp']
df['Family'].loc[df['Family'] > 0] = 1
df['Family'].loc[df['Family'] == 0] = 0

survival_analysis_of('Family')  
```

It's fair to say that if you were in a higher class you probably had a better chance of surviving. 

### 3.  Cherbourg, Queenstown, and Southampton were the locations passengers embarked from. We see that the marjoity of passengers embarked from Southampton. On average, the majority of users died from Southampton. The majority of passengers from Cherbourg survived followed by Queenstown.
